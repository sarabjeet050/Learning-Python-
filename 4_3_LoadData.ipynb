#### Introduction to Pandas

```python
# Import required library
import pandas as pd
```

###### After the import command, we now have access to a large number of pre-built classes and functions. This assumes the library is installed; in our lab environment all the necessary libraries are installed. One way pandas allow you to work with data is a dataframe. Let's go through the process to go from a comma separated values (<b>.csv</b>) file to a dataframe. This variable <code>csv_path</code> stores the path of the <b>.csv</b>, which is used as an argument to the <code>read_csv</code> function. The result is stored in the object <code>df</code>, this is a common short form used for a variable referring to a Pandas dataframe.


```python
# Read data from csv file
csv_path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/TopSellingAlbums.csv'
df = pd.read_csv(csv_path)
```

```python
# Print the first 5 rows of the dataframe
df.head()
```

```python
# Read the data from Excel and print the first 5 rows
xlsx_path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%204/Datasets/TopSellingAlbums.xlsx'
df = pd.read_excel(xlsx_path)
df.head()
```

```python
# Access the column Length and assign it to new dataframe x
x = df[['Length']] # double square brackets choose one or more columns as a dataframe.
x
```

#### Viewing and Accessing data

```python
# Get the column as a series. Think Pandas series as a 1-D dataframe. Just use 1 bracket
x = df['Length']
x
```

```python
# Get the column "Artist" as a dataframe
x = df[['Artist']] # double square brackets for dataframe
x
```

```python
# Access to multiple columns using multiple column headers enclosed in double bracket after dataframe name.
# Result is new dataframe comprised of specified columns

y = df[["Artist", "Length", "Genre"]]
y
```

###### One way to access unique elements is the "iloc" method.

```python
# Access the value on the first row and first column
df.iloc[0,0] # in iloc[row_no.,col_no.] # here row=0, colo=0
```

```python
# Access the value on the second row and the first column
df.iloc[1,0] # row=1, col=0
```

```python
# Access the value on the first row and the third column
df.iloc[0,2] #row=0, col=2
```

```python
# Access the value on the second row and the third column
df.iloc[1, 2] # row=1, col=2
```

###### Can also access the column using the name as well, the following is same as above:

```python
# Access the value on the first row and first column using name
df.loc[0, "Artist"]
```

```python
# Access the value on the second row and first column using name
df.loc[1, "Artist"]
```

```python
# Access the value on the first row and third column using name
df.loc[0, "Released"]
```

```python
# Access the value on the second row and third column using name
df.loc[1,"Released"]
```

#### Slicing
###### Can perform slicing using both the index and name of the column

```python
# iloc for index location and loc for labelled location
df.iloc[0:2, 0:3] # df.iloc[rowno_slice, colno_slice]; row_slice=0:2(0 to 1); col_slice=0:3(0 to 2)
```

```python
# slicing the dataframe using name
df.loc[0:2, "Artist": "Released"] # loc[rowlable_slice, collabel_slice]; slice here start:end both considered outcome
```
